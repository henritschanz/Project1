## B. Indexing and grouping

```python
# First we check what values of 'Region' exist in our data.
data3['Region'].unique()
```

```python
# We see that we have the extra region 'Europe', which we do not want. Let's check where
# it occurs in the data:
np.where(data3['Region']=='Europe')
data3[128:129]
```

```python
# Since it only occurs for the Ukraine, we can simply change this entry to 'Eastern Europe'
data3.loc[data3['Region'] == 'Europe', 'Region'] = 'Eastern Europe'
data3
```

```python
# Split the data according to Region
data3.groupby('Region')
```

```python
# Mean by Region
regional_means = data3.groupby('Region').mean()
regional_means['Happiness Score']
```

```python
# Rank the regions from most happy to least happy
regional_means.sort_values('Happiness Score', ascending=False)['Happiness Score']
```

```python
# Compute the number of countries in each region that have a happiness score above 6.0
data3[data3['Happiness Score'] > 6.0].groupby('Region').count()['Country']
```

```python
# Compute the difference between the maximum and minimum happiness score for each region
regional_max = data3.groupby('Region').max()['Happiness Score']
regional_min = data3.groupby('Region').min()['Happiness Score']
regional_range = regional_max - regional_min
regional_range
```

We can see that the largest range of happiness scores is found in the Asia-Pacific region, with a range of 3.852.

## C. Bar plot of the Happiness Score

```python
import matplotlib.pyplot as plt
%matplotlib inline

# Create a bar plot of the Happiness score of the ten happiest countries
x = np.arange(10)
y = data3.head(10)['Happiness Score']
countries = data3[0:10]['Country']
plt.figure(figsize=(8,4))
plt.barh(x,y)
plt.yticks(x,countries)
plt.xticks(np.arange(9))
plt.title('Happiness score of the ten happiest countries')
plt.xlabel('Happiness score')
plt.ylabel('Country')
plt.gca().invert_yaxis()
plt.show()
```

```python
# Create a detailed breakdown of the bar plot of the Happiness score of the ten happiest
# countries
plt.barh(x,y)
plt.yticks(x,countries)
plt.xticks(np.arange(9))
plt.title('Detailed Happiness score of the ten happiest countries')
plt.xlabel('Country')
plt.ylabel('Happiness score')

economy = data3[0:10]['Economy']
family = data3[0:10]['Family']
health = data3[0:10]['Health']
freedom = data3[0:10]['Freedom']
generosity = data3[0:10]['Generosity']
corruption = data3[0:10]['Corruption']
dystopia = data3[0:10]['Dystopia']

plt.barh(x, economy, label='Economy', color='orange')
plt.barh(x, family, left = economy, label='Family', color='pink')
plt.barh(x, health, left = economy+family, label='Health', color='magenta')
plt.barh(x, freedom, left = economy+family+health, label='Freedom', color='palegreen')
plt.barh(x, generosity, left = economy+family+health+freedom, label='Generosity', color='yellow')
plt.barh(x, corruption, left = economy+family+health+freedom+generosity, label='Corruption', color='brown')
plt.barh(x, dystopia, left = economy+family+health+freedom+generosity+corruption, label='Dystopia', color='grey')
plt.legend(bbox_to_anchor=(1,1))
plt.show()
```

```python
# Create a detailed breakdown of the bar plot of the Happiness score of the all african
# countries
african_data = data3[data3['Region']=='Africa']
new_x = np.arange(len(african_data))
new_y = african_data['Happiness Score']
plt.figure(figsize=(6,10))
plt.barh(new_x,new_y)
plt.yticks(new_x,african_data['Country'])
plt.xticks(np.arange(9))
plt.title('Detailed Happiness score of african countries')
plt.xlabel('Country')
plt.ylabel('Happiness score')

economy = african_data['Economy']
family = african_data['Family']
health = african_data['Health']
freedom = african_data['Freedom']
generosity = african_data['Generosity']
corruption = african_data['Corruption']
dystopia = african_data['Dystopia']

plt.barh(new_x, economy, label='Economy', color='orange')
plt.barh(new_x, family, left = economy, label='Family', color='pink')
plt.barh(new_x, health, left = economy+family, label='Health', color='magenta')
plt.barh(new_x, freedom, left = economy+family+health, label='Freedom', color='palegreen')
plt.barh(new_x, generosity, left = economy+family+health+freedom, label='Generosity', color='yellow')
plt.barh(new_x, corruption, left = economy+family+health+freedom+generosity, label='Corruption', color='brown')
plt.barh(new_x, dystopia, left = economy+family+health+freedom+generosity+corruption, label='Dystopia', color='grey')
plt.legend(bbox_to_anchor=(1,1))
plt.gca().invert_yaxis()
plt.show()
```

## H. Matrices

```python
# We define a matrix whose rows correspond to countries and the columns to the regions
num_countries = len(data)
num_regions = len(data['Region'].unique())

df = pd.DataFrame(np.zeros(num_countries*num_regions).reshape(num_countries,num_regions),
                  index = data['Country'], columns = data['Region'].unique())
for c in range(0,num_countries):
    region = data['Region'][c]
    country = data['Country'][c]
    df.loc[country,region] = 1
df
```
